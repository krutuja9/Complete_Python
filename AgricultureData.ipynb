```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

# Step 1: Clean the 'Cost of Activity' column to convert it into numeric format
def convert_to_numeric(value):
    try:
        return pd.to_numeric(value, errors='coerce')  # Convert to numeric, invalid parsing returns NaN
    except:
        return np.nan

# Apply the function to 'Cost of Activity' column
df['Cost of Activity'] = df['Cost of Activity'].apply(convert_to_numeric)

# Step 2: Separate rows with and without missing values in 'Cost of Activity'
df_not_missing = df[df['Cost of Activity'].notna()]
df_missing = df[df['Cost of Activity'].isna()]

# Step 3: Extract features and target
features = ['District', 'Crop', 'Varity', 'Week', 'Current Crop Stage', 'Type of Plot', 'Current Week Activity']
target = 'Cost of Activity'

# Step 4: One-hot encode categorical features
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
X_encoded = encoder.fit_transform(df_not_missing[features])

# Step 5: Impute any missing values in the feature columns
imputer = SimpleImputer(strategy='mean')
X_encoded = imputer.fit_transform(X_encoded)

# Step 6: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_encoded, df_not_missing[target], test_size=0.2, random_state=42)

# Step 7: Train a Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 8: Prepare the missing data for prediction
X_missing_encoded = encoder.transform(df_missing[features])
X_missing_encoded = imputer.transform(X_missing_encoded)

# Step 9: Predict the missing values for 'Cost of Activity'
df_missing['Cost of Activity'] = model.predict(X_missing_encoded)

# Step 10: Combine the datasets (rows with predicted and non-missing values)
df_filled = pd.concat([df_not_missing, df_missing])

# Display the resulting DataFrame with the filled values
print(df_filled)
```

```python
# Step 1: Set 'Labour Used' to 0 where 'Current Week Activity' is 'No Activity'
df_filled.loc[df_filled['Current Week Activity'] == 'No Activity', 'Labour Used'] = 0

# Step 2: Group by 'Current Week Activity' and fill 'Labour Used' with the mode
# First, define a function to fill the mode
def fill_labour_used_with_mode(group):
    # If there's no value in 'Labour Used', find the mode of the non-null values
    if group['Labour Used'].isnull().all():
        return group  # No value to fill, return group as is
    else:
        mode_value = group['Labour Used'].mode()
        if not mode_value.empty:  # Check if mode exists
            group['Labour Used'].fillna(mode_value[0], inplace=True)
    return group

# Step 3: Apply the mode-filling function to each group of 'Current Week Activity'
df_filled = df_filled.groupby('Current Week Activity', group_keys=False).apply(fill_labour_used_with_mode)

# Display the resulting DataFrame

df_filled
```

```python
# Step 1: Extract the last 2 characters from the 'Week' column and convert to integer
df_filled['Week'] = df_filled['Week'].str[-2:].astype(int)

# Display the updated DataFrame
df_filled
```
