```python
import matplotlib.pyplot as plt
import pandas as pd

data = {  "Cricket_Bat": ['sg','bdm','ss','gm','kookabura','spartan'],
    "MRP": [2000,2200,2400,2500,3000,3300],
    "Weight_Grams":[1100,1130,1200,1333,1480,1600]
}

df = pd.DataFrame(data)

# Plot a line using the pyplot,plt()
plt.plot(df["MRP"], df["Weight_Grams"])
plt.title("Bat Price depend on weight")
plt.xlabel("Bat Price (USD)")
plt.ylabel("Bat weight (Gram)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(5)
b = [2,4,6,8,10]
c = [1,3,5,7,9]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b, 'k--', label = "frequency")
ax.plot(a,c, 'k:', label="Periods")

ax.legend(loc= "upper center")


plt.title("Frequency of a signal")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(5)

b = [2,4,6,8,10]
c = [5,6,7,8,9]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b, 'k--', label = "frequency")
ax.plot(a,c, 'k:', label="Periods")

legend = ax.legend(loc= "upper center")

legend.get_frame().set_facecolor('red')

plt.title("Frequency of a signal")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

b = [2,4,6,8,10]
c = [5,6,7,8,9]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b , 'k--', label = "Frequency")
ax.plot(a,c , 'k:', label = "Periods")

ax.legend (loc= "upper center", fontsize = "xx-small")

plt.title("Frequency of a signal")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

student = np.array(["tim","jhon","amit","rk","karl"])
marks = np.array([87,56,66,77,99,])

plt.bar(student,marks)
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Student Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data
cricketer = np.array(["tim","jhon","amit","rk","karl","rkk"])
runs =  np.array([67,77,88,99,100,50])

plt.pie(runs, labels = cricketer, autopct= '%1.4f%%')

plt.title("runs scored in a cricket mathc")
plt.show()
```
